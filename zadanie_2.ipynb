```python
import pandas as pd
import matplotlib.pyplot as plt

prices = [
    (1, 2.12),
    (2, 2.56),
    (3, 3.10),
    (4, 3.16),
    (5, 3.58),
    (6, 5.12),
    (7, 5.16),
    (8, 5.20),
    (9, 4.12),
    (10, 4.10),
    (11, 3.65),
    (12, 4.25)
]

df = pd.DataFrame(prices, columns=["month", "price"])
df = df.set_index("month")
df
df['priceUSD'] = df['price'] / 4
df['priceUSD'].plot(color='red', linestyle='--') 
plt.title('Prices of goods (USD)')
plt.show()




```
